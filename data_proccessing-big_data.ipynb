### Now, we want to compare the data from OpenAlex with the data in arXiv. We're going to do this using OpenAlex API calls. Some calls return a 403 HTTP error code, so we'll clean up the lines that contain bad calls.

```python
import requests
from tqdm import tqdm


errors = []  # List to store error indexes
batch_size = 50
total_rows = len(arxiv_with_fixed_doi_df)
for i in tqdm(range(0, total_rows, batch_size)):
    dois = arxiv_with_fixed_doi_df['doi'].iloc[i:i + batch_size].tolist()
    pipe_separated_dois = "|".join(dois)

    r = requests.get(
        f"https://api.openalex.org/works?filter=doi:{pipe_separated_dois}&per-page=50&mailto=support@openalex.org")
    if r.status_code != 200:
        print(f'error indexes: {i}-{i + batch_size}')
        errors.append((i, i + batch_size))

        f = open("errors.txt", "a")
        f.write(f"error indexes: {i}-{i + batch_size}")
        f.close()

# Convert errors list to a DataFrame
errors_df = pd.DataFrame(errors, columns=['start_index', 'end_index'])
```

```python
# Extract the start and end indices
start_indices = errors_df['start_index'].tolist()
end_indices = errors_df['end_index'].tolist()

# Create a list of all indices
all_indices = []
for start, end in zip(start_indices, end_indices):
    indices = list(range(start, end + 1))  # Include both start and end indices
    all_indices.extend(indices)
```

```python
error_indices = []  # List to store the indices that cause errors

for i in all_indices:
    doi = arxiv_with_fixed_doi_df['doi'].iloc[i]
    try:
        url = f'https://api.openalex.org/works?filter=doi:{doi}'
        r = requests.get(url)
        r.raise_for_status()  # Raise exception for non-2xx response codes
        response_data = r.json()

    except requests.exceptions.HTTPError as err:
        print(f"HTTP Error: {err}")
        error_indices.append(i)  # Add the index to the error list

# Remove rows with error indices from temp_df
arxiv_with_fixed_doi_df.drop(error_indices, inplace=True)
```

### dois comparison between [OpenAlex](https://openalex.org/) and [arXiv dataset](https://www.kaggle.com/datasets/Cornell-University/arxiv) and extracting the relevant information

```python
splitted_csvs_folder ="splitted_csvs/"

if not os.path.exists(splitted_csvs_folder):
    os.makedirs(splitted_csvs_folder)
```

```python
def extract_relevant_data_by_doi_id(arxiv_doi, updated_works):
    for work in updated_works:
        if work['clean_doi_id'].lower() == arxiv_doi.lower():
            return work['cited_by_count'], work['publication_year'], work['publication_date'], work['type'], str(
                work['counts_by_year']), str(work['referenced_works']), work['id']
    return None, None, None, None, None, None, None  # Not exists

batch_size = 50
total_rows = len(arxiv_with_fixed_doi_df)
line_count = 0  # Counter for total lines

output_rows = []  # Accumulated rows to be saved in a CSV file

for i in tqdm(range(0, total_rows, batch_size)):
    dois = arxiv_with_fixed_doi_df['doi'].iloc[i:i + batch_size].tolist()
    pipe_separated_dois = "|".join(dois)
    r = requests.get(
        f"https://api.openalex.org/works?filter=doi:{pipe_separated_dois}&per-page=50&mailto=support@openalex.org")
    works = r.json()["results"]

    updated_works = [
        {**work, 'clean_doi_id': work['doi'].split('https://doi.org/')[1]}
        for work in works]

    (
        arxiv_with_fixed_doi_df.loc[i:i + batch_size - 1, 'total_citations_openalex'],
        arxiv_with_fixed_doi_df.loc[i:i + batch_size - 1, 'publication_year_openalex'],
        arxiv_with_fixed_doi_df.loc[i:i + batch_size - 1, 'publication_date_openalex'],
        arxiv_with_fixed_doi_df.loc[i:i + batch_size - 1, 'type_openalex'],
        arxiv_with_fixed_doi_df.loc[i:i + batch_size - 1, 'counts_by_year_openalex'],
        arxiv_with_fixed_doi_df.loc[i:i + batch_size - 1, 'referenced_works_openalex'],
        arxiv_with_fixed_doi_df.loc[i:i + batch_size - 1, 'id_openalex']
    ) = zip(*arxiv_with_fixed_doi_df['doi'].iloc[i:i + batch_size].apply(
        lambda doi: extract_relevant_data_by_doi_id(doi, updated_works)
    ))

    line_count += batch_size

    if line_count >= 100:
        start_line = (i + batch_size) - line_count
        end_line = i + batch_size

        arxiv_with_fixed_doi_df.iloc[start_line:end_line].to_csv(f'{splitted_csvs_folder}{start_line}-{end_line-1}_arxiv_with_openalex_data.csv', index=True)
        print(f'Saved rows {start_line} to {end_line - 1} to CSV')

        line_count = 0
```

### Unification of all CSVS into one updated CSV by OpenAlex and arXiv

```python
import os
import glob

# Get a list of all CSV files in the 'splitted_csvs_folder' folder
csv_files = glob.glob(os.path.join(splitted_csvs_folder, '*.csv'))

# Initialize an empty list to store the dataframes
dfs = []

# Iterate over each CSV file
for file in tqdm(csv_files):
    try:
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file)
        if not df.empty:
            # Append the DataFrame to the dfs list
            dfs.append(df)
    except pd.errors.EmptyDataError:
        print(f"Skipping empty file: {file}")

# Concatenate all dataframes in the list
combined_data = pd.concat(dfs, ignore_index=True)

# Write the combined_data DataFrame to a new CSV file
#combined_data.to_csv('arxiv_compared_with_openalex_with_ref_id.csv', index=False)
```
